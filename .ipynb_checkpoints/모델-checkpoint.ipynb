```python
import sys
from PyQt5.QtWidgets import *
from PyQt5 import uic
import PySide6.QtCore
from sklearn import preprocessing
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import pandas_datareader.data as web
from matplotlib import pyplot
import seaborn as sns
from sklearn.metrics import r2_score, mean_absolute_error
import joblib

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', None)

print("실행")

index = []
data = pd.read_csv("bangkok.csv")


for i in range(len(data)):
    if data['price'].iloc[i] == "매진":
        index.append(i)

data.drop(index = index, axis=0, inplace=True)
data = data.fillna("X")

for i in range(9089):
    
    data['price'].iloc[i] = data['price'].iloc[i][3:12]
    data['price'].iloc[i] = data['price'].iloc[i].replace(",", "")
    data['price'].iloc[i] = data['price'].iloc[i].replace("원", "")
    data['price'].iloc[i] = data['price'].iloc[i].replace(" ", "")
    data['price'].iloc[i] = int(data['price'].iloc[i])
    if  data['price'].iloc[i] > 2000000:
        data['price'].iloc[i] = 2000000
        
    data['date'].iloc[i] = data['date'].iloc[i].replace("월 ", "-")
    data['date'].iloc[i] = data['date'].iloc[i].replace("일", "")
    
    data['duration'].iloc[i] = data['duration'].iloc[i][:2]
    if data['duration'].iloc[i][1] == "시":
        data['duration'].iloc[i] = data['duration'].iloc[i].replace("시", "")
    data['duration'].iloc[i] = int(data['duration'].iloc[i])
    
    data['time1'].iloc[i] = str(data['time1'].iloc[i])
    data['time1'].iloc[i] = data['time1'].iloc[i][:2]
    if data['time1'].iloc[i][1] == ":":
        data['time1'].iloc[i] = data['time1'].iloc[i].replace(":", "")
    data['time1'].iloc[i] = int(data['time1'].iloc[i])
    if 0 <= data['time1'].iloc[i] and data['time1'].iloc[i] <= 6:
        data['time1'].iloc[i] = "새벽"
    elif 6 < data['time1'].iloc[i] and data['time1'].iloc[i] <= 12:
        data['time1'].iloc[i] = "오전"
    else:
        data['time1'].iloc[i] = "오후"
        
    data['time2'].iloc[i] = str(data['time2'].iloc[i])
    data['time2'].iloc[i] = data['time2'].iloc[i][:2]
    if data['time2'].iloc[i][1] == ":":
        data['time2'].iloc[i] = data['time2'].iloc[i].replace(":", "")
    data['time2'].iloc[i] = int(data['time2'].iloc[i])
    if 0 <= data['time2'].iloc[i] and data['time2'].iloc[i] <= 6:
        data['time2'].iloc[i] = "새벽"
    elif 6 < data['time2'].iloc[i] and data['time2'].iloc[i] <= 12:
        data['time2'].iloc[i] = "오전"
    else:
        data['time2'].iloc[i] = "오후"
    
    try:
        data['transfer'].iloc[i] = str(data['transfer'].iloc[i])
        if len(data['transfer'].iloc[i]) != 3:
            data['transfer'].iloc[i] = "O"
        else:
            data['transfer'].iloc[i] = "X"
            
    except:
        data['transfer'].iloc[i] = "X"

data = data.astype({'duration':'int64', 'price':'int64'})
data = data.drop(['transfer'], axis=1)

le = preprocessing.LabelEncoder()
data['time1'] = le.fit_transform(data['time1'])
data['time2'] = le.fit_transform(data['time2'])
data['flight_number1'] = le.fit_transform(data['flight_number1'])
data['flight_number2'] = le.fit_transform(data['flight_number2'])
data['country'] = le.fit_transform(data['country'])
data['date'] = le.fit_transform(data['date'])

Y = data["price"]
X = data.loc[:, data.columns != 'price']
validation_size = 0.2
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=validation_size, random_state=seed)

model = joblib.load('filename.pkl') 

form_class = uic.loadUiType("untitled.ui")[0]

data['date'] = le.inverse_transform(data['date'])

class MyWindow(QMainWindow, form_class):

    def __init__(self):
        super().__init__()
        self.setupUi(self)
        self.pushButton.clicked.connect(self.btn_clicked)
        
    def btn_clicked(self):
        if self.comboBox.currentText() == "방콕":
            country = 0
        elif self.comboBox.currentText() == "괌":
            country = 1
        elif self.comboBox.currentText() == "하와이":
            country = 2
        elif self.comboBox.currentText() == "뉴욕":
            country = 3
        elif self.comboBox.currentText() == "도쿄":
            country = 4
            
        date = self.dateEdit.date()
        date = date.toString()
        month = date[2]
        day = date[4]
        if int(day) >= 10:
            date2 = "0" + month + "-" + day
        else:
            date2 = "0" + month + "-" + "0" + day
        print(date2, country)
        index2 = 0
        flag = False
        
        text = QTableWidgetItem(self.comboBox.currentText())
        date = QTableWidgetItem(date2)
        
        self.tableWidget.setItem(0, 1, text)
        self.tableWidget.setItem(0, 2, date)
        
        for i in range(9089):
            if data['date'].iloc[i] == date2 and data['country'].iloc[i] == country:
                flag = True
                index2 = i    
            
        if flag:
            data['date'] = le.fit_transform(data['date'])
            price = model.predict(X)
            self.tableWidget.setItem(0, 0, QTableWidgetItem(str(price[index2])))
            print(price[index2])
        else:
            self.tableWidget.setItem(0, 0, QTableWidgetItem("매진"))
            print("매진")
        
        
app = QApplication(sys.argv)
window = MyWindow()
window.show()
app.exec_()
```
